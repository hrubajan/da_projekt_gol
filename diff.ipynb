```python
stops_vs_positions_clean = pd.read_csv('stops_vs_positions_clean.csv', encoding="utf-8", low_memory=False)
stops_vs_positions_clean.head()
```

```python
final = stops_vs_positions_clean
final[final['UNICORN']== '331_241_200302_06_May']
```

```python
unicorn_list = final['UNICORN'].unique().tolist()
len(unicorn_list)
```

```python
for uni in unicorn_list:
    
    unicorn_find = final[final['UNICORN'] == uni]
    
    differences = unicorn_find[['DELAY']].diff(periods=1)
    
    print(differences)

    for i,diff in enumerate(differences['DELAY']):
        final.loc[i,'DIFF_DELAY'] = diff
```
